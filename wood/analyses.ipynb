```python
import numpy as np
import pandas as pd
import os

path = '/Users/bilyanavencheva/IdeaProjects/ucdp-smolian-scraper/exports/excel/clean_all'
pd.options.display.float_format = '{:,.0f}'.format
```

```python
clean_all = None
for file in os.listdir(path):
    if '.xlsx' in file:
        df = pd.read_excel(os.path.join(path,file))
    else:
        pass
    
    clean_all = df if clean_all is None else pd.concat([clean_all,df],ignore_index=True)
```

```python
clean_all.columns
```

```python
cena = 'Договорена цена (лв.)' #'Начална цена'
sum_ = clean_all.groupby(['ДП','година']).sum()[[cena,'Обем дървесина (м3)']]
for col in sum_.columns:
    sum_.rename(columns ={col:f'{col}_(общо)'},inplace = True)

count_ = clean_all.groupby(['ДП','година']).count()[['дата']].rename(columns = {'дата':'брой'})

mean_ = clean_all.groupby(['ДП','година']).mean()[[cena,'Обем дървесина (м3)']]
for col in mean_.columns:
    mean_.rename(columns ={col:f'{col}_(средно)'},inplace = True)


grouped_detailed = count_.merge(sum_, left_index =True, right_index = True).merge(mean_, left_index =True, right_index = True)

grouped_detailed
```

```python
sum_ = clean_all.groupby(['ДП']).sum()[[cena,'Обем дървесина (м3)']]
for col in sum_.columns:
    sum_.rename(columns ={col:f'{col}_(общо)'},inplace = True)

count_ = clean_all.groupby(['ДП']).count()[['дата']].rename(columns = {'дата':'брой'})

mean_ = clean_all.groupby(['ДП']).mean()[[cena,'Обем дървесина (м3)']]
for col in mean_.columns:
    mean_.rename(columns ={col:f'{col}_(средно)'},inplace = True)


grouped_ = count_.merge(sum_, left_index =True, right_index = True).merge(mean_, left_index =True, right_index = True)
grouped_
```
